4. Как найти самое часто встречаемое значение между двумя массивами? 

````python
import numpy as np

def most_common_between_arrays(arr1, arr2):
    combined = np.concatenate([arr1, arr2])
    unique, counts = np.unique(combined, return_counts=True)
    max_count_idx = np.argmax(counts)
    return unique[max_count_idx], counts[max_count_idx]

arr1 = np.array([1, 2, 3, 4, 2, 2])
arr2 = np.array([2, 3, 3, 5, 1])
value, count = most_common_between_arrays(arr1, arr2)
print(f"Самое частое значение: {value}, встречается {count} раз")
````

Набор данных описывает эффективность учеников в зависимости от некоторых факторов влияющих на обучение и окружение человека.

Файл: `student_perfomance.csv`

```
Hours_Studied - количество часов затраченное на обучение в неделю
Attendance - поцент посещенных занятий
Parental_Involvement - уровень участия родителей в обучении учащегося
Access_to_Resources - доступность ресурсов для обучения
Extracurricular_Activities - участие во внеклассных мероприятиях
Sleep_Hours - количество часов выделяемое на сон в день
Previous_Scores - оценка за первый экзамен
Motivation_Level - уровень мотивации
Internet_Access - доступ к интернету
Tutoring_Sessions - количество посещаемых репетиторских занятий в месяц.
Family_Income - доход семьи
Teacher_Quality - насколько качественно преподает учитель
School_Type - тип школы(частная, обычная)
Peer_Influence - влияние сверстников
Physical_Activity - физическая активность
Learning_Disabilities - наличие проблем с обучаемостью
Parental_Education_Level - уровень образования родителей
Distance_from_Home - дистанция от дома до школы
Gender - пол
Exam_Score - оценка на конечном экзамене
```

````python
import pandas as pd
import numpy as np

np.random.seed(42)

n_students = 1000

data = {
    'Hours_Studied': np.random.normal(15, 5, n_students).clip(5, 30),
    'Attendance': np.random.normal(85, 10, n_students).clip(50, 100),
    'Parental_Involvement': np.random.choice([1, 2, 3, 4, 5], n_students, p=[0.1, 0.2, 0.4, 0.2, 0.1]),
    'Access_to_Resources': np.random.choice([1, 2, 3, 4, 5], n_students, p=[0.05, 0.15, 0.3, 0.3, 0.2]),
    'Extracurricular_Activities': np.random.choice(['Нет', 'Редко', 'Иногда', 'Часто'], n_students, p=[0.2, 0.3, 0.3, 0.2]),
    'Sleep_Hours': np.random.normal(7.5, 1.2, n_students).clip(4, 10),
    'Previous_Scores': np.random.normal(75, 12, n_students).clip(40, 95),
    'Motivation_Level': np.random.choice([1, 2, 3, 4, 5], n_students, p=[0.1, 0.15, 0.25, 0.3, 0.2]),
    'Internet_Access': np.random.choice(['Да', 'Нет'], n_students, p=[0.9, 0.1]),
    'Tutoring_Sessions': np.random.poisson(2, n_students).clip(0, 8),
    'Family_Income': np.random.choice(['Низкий', 'Средний', 'Высокий'], n_students, p=[0.3, 0.5, 0.2]),
    'Teacher_Quality': np.random.choice([1, 2, 3, 4, 5], n_students, p=[0.05, 0.15, 0.4, 0.3, 0.1]),
    'School_Type': np.random.choice(['Частная', 'Обычная'], n_students, p=[0.3, 0.7]),
    'Peer_Influence': np.random.choice(['Отрицательное', 'Нейтральное', 'Положительное'], n_students, p=[0.2, 0.5, 0.3]),
    'Physical_Activity': np.random.choice(['Низкая', 'Умеренная', 'Высокая'], n_students, p=[0.3, 0.5, 0.2]),
    'Learning_Disabilities': np.random.choice(['Да', 'Нет'], n_students, p=[0.1, 0.9]),
    'Parental_Education_Level': np.random.choice(['Среднее', 'Среднее специальное', 'Высшее'], n_students, p=[0.4, 0.3, 0.3]),
    'Distance_from_Home': np.random.exponential(2, n_students).clip(0.1, 10),
    'Gender': np.random.choice(['Мужской', 'Женский'], n_students, p=[0.48, 0.52])
}

df = pd.DataFrame(data)

def calculate_exam_score(row):
    base_score = 50
    
    score = (base_score +
             row['Hours_Studied'] * 0.4 +
             row['Attendance'] * 0.15 +
             row['Parental_Involvement'] * 1.5 +
             row['Access_to_Resources'] * 1.2 +
             {'Нет': 0, 'Редко': 0.5, 'Иногда': 1, 'Часто': 0.5}[row['Extracurricular_Activities']] +
             (row['Sleep_Hours'] - 6) * 0.8 +
             row['Previous_Scores'] * 0.2 +
             row['Motivation_Level'] * 2 +
             (1 if row['Internet_Access'] == 'Да' else -2) +
             row['Tutoring_Sessions'] * 1.2 +
             {'Низкий': -2, 'Средний': 0, 'Высокий': 2}[row['Family_Income']] +
             row['Teacher_Quality'] * 1.5 +
             (3 if row['School_Type'] == 'Частная' else 0) +
             {'Отрицательное': -3, 'Нейтральное': 0, 'Положительное': 2}[row['Peer_Influence']] +
             {'Низкая': -1, 'Умеренная': 1, 'Высокая': 2}[row['Physical_Activity']] +
             (-8 if row['Learning_Disabilities'] == 'Да' else 0) +
             {'Среднее': 0, 'Среднее специальное': 1, 'Высшее': 2}[row['Parental_Education_Level']] +
             -row['Distance_from_Home'] * 0.3 +
             (1 if row['Gender'] == 'Женский' else 0))
    
    score += np.random.normal(0, 5)
    
    return max(0, min(100, score))

df['Exam_Score'] = df.apply(calculate_exam_score, axis=1)

numeric_columns = ['Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores', 
                  'Tutoring_Sessions', 'Distance_from_Home', 'Exam_Score']
df[numeric_columns] = df[numeric_columns].round(2)

df.to_csv('dataset.csv', index=False, encoding='utf-8')

print(f"Датасет успешно создан! Размер: {df.shape}")
````

5. Правда ли, что те, кто спит больше(75 перцентиль и выше) сдает экзамен лучше, чем студенты с недосыпом(25 перцентиль и меньше)

````python
import pandas as pd
from scipy import stats

df = pd.read_csv('dataset.csv')

sleep_25 = df['Sleep_Hours'].quantile(0.25)
sleep_75 = df['Sleep_Hours'].quantile(0.75)

print(f"25-й перцентиль сна: {sleep_25:.2f} часов")
print(f"75-й перцентиль сна: {sleep_75:.2f} часов")

well_rested = df[df['Sleep_Hours'] >= sleep_75]
sleep_deprived = df[df['Sleep_Hours'] <= sleep_25]

print(f"\nКоличество студентов:")
print(f"- Высыпающиеся (≥{sleep_75:.2f} ч): {len(well_rested)}")
print(f"- Недосыпающие (≤{sleep_25:.2f} ч): {len(sleep_deprived)}")

well_rested_avg = well_rested['Exam_Score'].mean()
sleep_deprived_avg = sleep_deprived['Exam_Score'].mean()

print(f"\nСредние баллы экзаменов:")
print(f"- Высыпающиеся: {well_rested_avg:.2f}")
print(f"- Недосыпающие: {sleep_deprived_avg:.2f}")
print(f"Разница: {well_rested_avg - sleep_deprived_avg:.2f} балла")

t_stat, p_value = stats.ttest_ind(well_rested['Exam_Score'], sleep_deprived['Exam_Score'])

print(f"\nРезультат t-теста:")
print(f"t-статистика: {t_stat:.4f}")
print(f"p-значение: {p_value:.4f}")

if p_value < 0.05:
    if well_rested_avg > sleep_deprived_avg:
        print("\nДА, студенты, которые спят больше, сдают экзамены статистически значимо лучше")
    else:
        print("\nНЕТ, студенты с недосыпом сдают экзамены статистически значимо лучше")
else:
    print("\nНет статистически значимой разницы в результатах экзаменов")
````

6. Влияет ли физическая активность на результаты экзамена?

````python
import pandas as pd
import scipy.stats as stats
from scipy.stats import f_oneway

df = pd.read_csv('dataset.csv')

physical_activity_groups = df.groupby('Physical_Activity')['Exam_Score']

activity_stats = physical_activity_groups.agg(['count', 'mean', 'std', 'median'])
print("Статистика по группам физической активности:")
print(activity_stats)

print(f"\nСредние баллы по уровням физической активности:")
for activity in df['Physical_Activity'].unique():
    group_data = df[df['Physical_Activity'] == activity]['Exam_Score']
    print(f"- {activity}: {group_data.mean():.2f} ± {group_data.std():.2f} (n={len(group_data)})")

groups = [df[df['Physical_Activity'] == activity]['Exam_Score'] 
          for activity in df['Physical_Activity'].unique()]

f_stat, p_value = f_oneway(*groups)

if p_value < 0.05:
    print("\nДА, уровень физической активности статистически значимо влияет на результаты экзамена")
else:
    print("\nНЕТ, уровень физической активности не оказывает статистически значимого влияния на результаты экзамена")
````

10. Посчитайте долю студентов сдавших экзамен на более чем 70 баллов в семьях с низким доходом и отсутствием доступа к интернету

````python
import pandas as pd

df = pd.read_csv('dataset.csv')

filtered_students = df[
    (df['Family_Income'] == 'Низкий') & 
    (df['Internet_Access'] == 'Нет')
]

print("Критерии отбора:")
print("- Семейный доход: Низкий")
print("- Доступ к интернету: Нет")

total_filtered = len(filtered_students)
print(f"\nОбщее количество студентов с низким доходом и без интернета: {total_filtered}")

passed_70_plus = filtered_students[filtered_students['Exam_Score'] > 70]
count_passed = len(passed_70_plus)

print(f"Количество студентов, сдавших на >70 баллов: {count_passed}")

if total_filtered > 0:
    percentage = (count_passed / total_filtered) * 100
    print(f"Доля студентов, сдавших на >70 баллов: {percentage:.1f}%")
else:
    print("Нет студентов, удовлетворяющих критериям отбора")

if total_filtered > 0:
    if count_passed > 0:
        print(f"\nСтатистика для студентов с баллами >70:")
        print(f"Средний балл (>70): {passed_70_plus['Exam_Score'].mean():.2f}")
        print(f"Медианный балл (>70): {passed_70_plus['Exam_Score'].median():.2f}")
    
    excellent = len(filtered_students[filtered_students['Exam_Score'] >= 90])
    good = len(filtered_students[(filtered_students['Exam_Score'] >= 70) & (filtered_students['Exam_Score'] < 90)])
    below_70 = len(filtered_students[filtered_students['Exam_Score'] < 70])
    
    print(f"\nРаспределение по уровням успеваемости:")
    print(f"- Отлично (≥90): {excellent} студентов ({(excellent/total_filtered)*100:.1f}%)")
    print(f"- Хорошо (70-89): {good} студентов ({(good/total_filtered)*100:.1f}%)")
    print(f"- Ниже 70: {below_70} студентов ({(below_70/total_filtered)*100:.1f}%)")
````
